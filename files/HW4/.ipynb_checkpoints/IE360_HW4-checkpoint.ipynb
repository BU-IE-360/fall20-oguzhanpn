```python
import pandas as pd
import numpy as np
import seaborn as sns 
import datetime as dt
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.seasonal import DecomposeResult, seasonal_decompose
from statsmodels.tsa.stattools import kpss
import warnings
warnings.filterwarnings("ignore")
import pmdarima as pm
```

# 1. Read data

```python
# read the data
train = pd.read_excel(r"GercekZamanliTuketim-01012017-08012021.xls")
test = pd.read_excel(r"GercekZamanliTuketim-09012021-23012021.xls")

# fix the column names, convert to dates to datetime and convert consumption values to float
train.columns = ["date", "hour", "consumption"]
test.columns = ["date", "hour", "consumption"]

train.date = train.date.apply(lambda x: dt.datetime.strptime(x, "%d.%m.%Y"), 1)
test.date = test.date.apply(lambda x: dt.datetime.strptime(x, "%d.%m.%Y"), 1)

train.consumption = train.consumption.apply(lambda x: x.replace(".", "").replace(",", ".")).astype(float)
test.consumption = test.consumption.apply(lambda x: x.replace(".", "").replace(",", ".")).astype(float)
```

Since my data is hourly I will calculate the daily cansumption values by summing the days.

```python
train = pd.DataFrame(train.groupby(by = "date").consumption.sum())
test = pd.DataFrame(test.groupby(by = "date").consumption.sum())
```

# 3. Decompose the Trend and Seasonality

In order to have a stationarity time series. I will first decompose the trend and seasonality component from the electricity consumption series. Than, I will analyze the big residuals in order to clean my data further.

```python
dcmp = seasonal_decompose(train.consumption, model = "additive", period=7)
#Set the figure size: 
sns.set(rc={'figure.figsize':(19,19)})

#Divide the figure:
fig, axs = plt.subplots(3,1)
fig.suptitle('Decomposition results', y = 0.92)

#Plot each variable with frgn_deposit_share
sns.lineplot(data = dcmp.trend, ax = axs[0])
sns.lineplot(data = dcmp.seasonal, ax = axs[1])
sns.lineplot(data = dcmp.resid, ax = axs[2])
```

Now, I will analyze the big residuals.

```python
print("The mean of the residuals is", dcmp.resid.mean(), "and the standart deviation is", dcmp.resid.std(), ". Now I will list the residuals behind 1.96 std deviation to analyze the reasons behind them.")
```

```python
my_residuals = pd.DataFrame(dcmp.resid[(dcmp.resid < (dcmp.resid.mean() - 1.96*dcmp.resid.std())) | (dcmp.resid > (dcmp.resid.mean() + 1.96*dcmp.resid.std()))])
my_residuals
```

When we analyze the residuals, we see that, most of them are resulting from the national or religious holidays. Since there is no such date in our test data (9 Jan - 23 Jan), I can get rid of those values by changing them with the mean of the weekday.

```python
train["weekday"] = train.index.weekday
mean_cons = train.groupby("weekday").mean()
for i in my_residuals.index: 
    train.loc[i, "consumption"] = mean_cons.loc[train.loc[i,"weekday"], "consumption"]
```

Now to be sure about stationarity I will check rolling mean, rolling variance and rolling autocorrelation

```python
sns.set(rc={'figure.figsize':(19,19)})

#Divide the figure:
fig, axs = plt.subplots(3,1)
fig.suptitle('rolling mean variance and autocorrelatioin', y = 0.92)

#Plot each variable with frgn_deposit_share

sns.lineplot(data=dcmp.resid.rolling(window=7).mean(), ax = axs[0])
sns.lineplot(data=(dcmp.resid.rolling(window=7)).var(), ax = axs[1])
sns.lineplot(data=(dcmp.resid.rolling(window=7)).corr(dcmp.resid), ax = axs[2])
```

```python
def kpss_test(timeseries):
    print ('Results of KPSS Test:')
    kpsstest = kpss(timeseries, regression='ct')
    kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
    for key,value in kpsstest[3].items():
        kpss_output['Critical Value (%s)'%key] = value
    print (kpss_output)
    
kpss_test(train.consumption)
```

we can say that we have reached the stationarity since the main features of our series doesn't change over time. The only problem is seem during the COVID period. But it is not a huge problem. Also kpss test shows that the series is stationary with p value 0.1. 

# 4. Modelling

```python
X= train.drop("consumption",1)
y = train.consumption
```

```python
# Fit your model
model = pm.auto_arima(y=y, seasonal=True, m=7)
model
```

The best parameters for ARIMA model is p,d,q = 3,0,0. Where,  

        p: the lag order (number of lag observations included)        
        d: the degree of differencing needed for stationarity (number of times the data is differenced)
        q: the order of the moving average

```python
forecasts = model.predict(test.shape[0])  

daily_reports = pd.DataFrame()
for i,date in enumerate(test.index):
    daily_reports.loc[date,"daily_bias"] = (test.consumption[i]- forecasts[i])
    daily_reports.loc[date,"daily_MAPE"] = abs(test.consumption[i]- forecasts[i])/test.consumption[i]
    
daily_reports
```

```python
def perf_dt(actual,forecast):
    n=len(actual)
    error=actual-forecast
    my_mean=actual.mean()
    my_sd = actual.std()
    FBias=sum(error)/sum(actual)
    MPE=sum(error/actual)/n
    MAPE=sum(abs(error/actual))/n
    RMSE= (sum(error*error)**(1/2))/n
    MAD=sum(abs(error))/n
    WMAPE=MAD/my_mean
    l={"n":n,"mean":my_mean,"std":my_sd,"Fbias":FBias,"MAPE":MAPE,"RMSE":RMSE,"MAD":MAD,"WMAPE":WMAPE}
    return(l)

perf_dt(test.consumption, forecasts)
```

We have WMAPE score of 0.07 it is not a perfect score, but it is acceptable. We can improve our model by adding some exogenous variables but it is beyond the content of this homework.
